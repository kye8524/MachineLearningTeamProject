### 이름에 크리스마스, 캐럴이 포함된 플레이스트  개수, 목록

```python
playlist = []
for index in train_data.index:
    tag = train_data['plylst_title'][index]
    if('크리스마스' in tag or '캐럴' in tag or '캐롤' in tag):
        playlist.append(index)
#         print(tag)
#             print("{:8} 제목 : ".format(index)+ df_train['plylst_title'][index])
print(len(playlist))
```

```python
for index in playlist:
    print("index : {:}".format(train_data['plylst_title'][index]))
```

```python
#위 크리스마스 플레이리스트 속 노래는 몇개? -> 라벨링 전 : 0   후 : 15050
carolNum = 0
isCarol = False
for song in song_meta.index:
    isCarol = False
    for genre in song_meta['song_gn_gnr_basket'][song]:
        if genre == "GN3000":
            isCarol = True;
            break
    if isCarol == True:
        carolNum += 1
        # print("Artist : {:}".format(df_song['artist_name_basket'][song]))
        # print("song : {:}".format(song_meta['song_name'][song]))
print("the number of carol : {:}".format(carolNum))
```

```python
#carol 리스트에 캐롤 추가해주기   - 한번만 돌리기
carol_list = []
for index in playlist:  # 크리스마스 플레이리스트 번호 돌기
    for songs in train_data['songs'][index]:
        if songs not in carol_list:
            carol_list.append(songs)
```

```python
#캐롤 리스트의 노래에 장르 추가해주기  - 한번만 돌리기 (라벨 추가)
for song in carol_list:
    song_meta['song_gn_gnr_basket'][song].append("GN3000")
```

```python
# #train 줄이기    -> 캐롤만 있는 플레이리스트 찾기
playlist_train = pd.DataFrame([], columns=['Index', 'tags', 'id', 'plylst_title', 'songs', 'like_cnt', 'updt_date'])
isInCarol = False
i=0;
for row in train_data.itertuples():
    isInCarol = False
    for song in train_data['songs'][row.Index]:
        if song in carol_list :
            isInCarol = True
            break
    if isInCarol == True :
        i += 1
        playlist_train.loc[i] = row
        if i>= 20000:
            break
```
